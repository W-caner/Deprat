## 评价指标
以上均为从【对比】的角度，以原api文档为基线，但原api也是人为手工书写的，下面将自己定义的评价指标SPDE（Standard Parameter Description Evaluation）。

该指标主要针对模型生成的第一句话进行评价：因为api参数描述可能是多方面的，例如“类型，可选，默认值等”，而这些应该是由规则生成，而非模型生成的，所以我们只关注某一个句子（往往是第一个），作为对该参数的意义描述。称之为“MPD（Meaningful Parameter Description）”。
- 【直接性】和【有意义性】：我们期望生成的语句是有意义的，且位置越靠前越直接。比如第一句话是“the name of the branch”将比“optional parameter”更有意义。
    1. 参数名称一般具有直接意义，将【常量命名/驼峰命名/帕斯卡命名/下划线命名】进行划分，名称提及率作为评估因素之一。
    2. 名称提及率过低时，此时将后续句子作为MPD。
    3. 随着MPD出现的位置往后，其直接性越差，最好的直接性为1，最差的直接性为len(description.split("."))。
    4. 将名称在MPD中的提及率作为有意义性，最好的有意义性为1，最差的有意义性为0。
- 【可读性】：我们倾向于完整且简洁的描述，有意义性可以衡量是否完整，将MPD中出现的有意义词汇率作为可读性。
    - 可读性低说明句子过长。
    - 重复词汇予以惩罚。
- 【一致性】：我们期望同一个文档尽量使用同一种描述方式，比如“the name of the branch”和“branch name”将被视为两种描述方式。

【直接性】【有意义性】和【可读性】都是对于某个生成的参数描述而言。而【一致性】则是针对整个api文档的规范。

```python
import re
import json
import tqdm
from collections import defaultdict, Counter

with open('D:\\王灿2190830726\\02_王灿_2190830726_ZX_CL\\GPT\\dataset\\test.json','rb') as f:
    rawdata = f.readlines()
names = []
infs = []
for data in tqdm.tqdm(rawdata):
    testdata = json.loads(data)
    inf = testdata['information']
    names.append(inf[inf.rfind("#name:")+6: inf.find("#type")])
    infs.append(inf)
    
# names[:50]

def div_name(name):
    # 常量命名/驼峰命名/帕斯卡命名 ->下划线命名
    _name = re.sub('(?<=[a-z])[A-Z]|(?<!^)[A-Z](?=[a-z])', '_\g<0>', name).lower()
    # 下划线命名
    return _name.strip("$").replace("-","_").split("_")
```

```python
def cal_MPDs(labels):
    MPDs = []
    score, consist_score = 0, 0
    mean_scores, direct_scores, read_scores = 0, 0, 0
    for i,label in enumerate(labels):
        divname = div_name(names[i])
        MPDs_pre = re.split('[:.;!?]', label)
        mean_score, direct_score, read_score = 0, 0, 0
        for idx, MPD in enumerate(MPDs_pre):
            # 剔除重复的话 
            if (sorted(Counter(MPD.split(" ")).items(),key=lambda x:x[1],reverse=True)[0][1] > 5):
                continue                
            # 根据名称提及率提取MPD
            cnt_mention = 0
            for word in divname:
                if word in MPD.replace(" ",""):
                    cnt_mention += 1
            tmp = cnt_mention/len(divname)
            # 等于的情况不往后
            if (tmp > mean_score):
                # 计算直接性，有意义性，可读性
                mean_score = cnt_mention/len(divname)
                direct_score = idx
                # direct_score = (idx+1)/len(MPD_pre)
                read_score = cnt_mention/len(MPD.split(" "))
                # break
        MPDs.append(MPDs_pre[direct_score])
        # 有意义性为0的情况直接性往后，MPD不往后
        if (mean_score==0):
            direct_score = len(MPDs_pre)
        if (read_score==1):
            read_score = 0
        # 根据重复词汇修正可读性，可能出现负数
        # read_score -= max(Counter(MPD.split(" ")).items())/len(MPD.split(" "))
        mean_scores +=mean_score
        direct_scores+=direct_score
        read_scores+=read_score
    print(direct_scores/len(labels), mean_scores/len(labels), read_scores/len(labels))
    return MPDs
```

```python
from stanfordcorenlp import StanfordCoreNLP

nlp = StanfordCoreNLP(r'D:/stanford-corenlp-4.5.3')

def classifer(sentence= 'enddatetime'):
    # print(nlp.word_tokenize(sentence))
    # print(nlp.pos_tag(sentence))
    # print(nlp.ner(sentence))
    if not sentence.strip(" \n"):
        return "None"
    if (sorted(Counter(sentence.split(" ")).items(),key=lambda x:x[1],reverse=True)[0][1] > 5):
        sentence = " ".join(sentence.split(" ")[:40])  
    tree = nlp.parse(sentence.replace('"',"").replace("'","").replace('/',"").replace("`","").replace(")","").replace("(","").replace("*",""))
    # nlp.dependency_parse(sentence)
    # 剔除重复的话 
    stack = []
    layers = defaultdict(list)
    layer = 0
    beginlayer = 2e5
    while tree:
        tree = tree.strip(' \r\n')
        if tree[0]=='(':
            dividx = tree.strip(' (').find('(') +1
            if (dividx==0) or (')' in tree[:dividx]):
                dividx = tree.strip('(').find(')') +1
            stack.append((layer,tree[:dividx].strip("\n\r '")))
            layer+=1
            tree = tree[dividx:]
        elif tree[0]==')':
            item = stack.pop()
            # （层数，词性，【单词】）
            l = item[0]
            p_and_w = item[1].strip('(').split(" ")
            if len(p_and_w)>1:
                beginlayer = min(beginlayer, l)
            layers[l].append(p_and_w[0].strip(" "))
            layer-=1
            tree = tree[1:]
    # 句式解析
    res = None
    for idx in range(beginlayer,0,-1):
        # print(layers[idx])
        # if layers[idx] == ['NP', 'PP'] or layers[idx] == ['NP', 'PP', 'VP']:
        if layers[idx] == ['NP', 'PP'] or layers[idx] == ['NP', 'S'] or layers[idx] == ['NP','NP','VP'] or layers[idx]==['NP-TMP','NP','VP']:
            res = 'NP+PP'
            break
            # return res
        elif layers[idx] == ['NP', 'VP']:
            res = 'NP+VP'
            break
            # return res
        elif layers[idx]==['NP'] or set(layers[idx]) == set(['NP','NP']):
            res = 'NP'
            break
            # return res
        elif layers[idx]==['VP']:
            res = 'VP'
            break
            # return res
        # elif layers[idx] == ['NP','S']:
            # res = 'NP+S'
            # return res
        # elif layers[idx]==['FRAG']:
            # res = 'FRAG'
            # return res
        elif 'SBAR' in layers[idx] or 'SBARQ' in layers[idx]:
            res = 'SBAR'
            break
            # return res 
    # print(sentence+'\n', layers)
    otherres = layers[2]
    while ',' in otherres:
        otherres.remove(',')
    return res if res else str(otherres)
```
